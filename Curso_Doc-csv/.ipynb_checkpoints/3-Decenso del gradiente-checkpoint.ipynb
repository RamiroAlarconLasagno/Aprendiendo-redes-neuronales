# Breve explicación:
##### El método del descenso del gradiente es un algoritmo de optimización que permite converger hacia el valor mínimo de una función mediante un proceso iterativo. En aprendizaje automático básicamente se utiliza para minimizar una función que mide el error de predicción del modelo en el conjunto de datos.
En este ejemplo tomaremos una función y una coordenada, a partir de esa coordenada se recorrera la función con el fin de terminar en la coordenada con el menor valor de la función. Para esto se creará la función, se generara una coordenada al azar y luego utilizando el metodo del descenso del gradiente el algoritmo recorrera el plano buscando el menor valor.

# Librerias:

```python
import numpy as np
import scipy as sc

import matplotlib.pyplot as plt
```

# Función:
## Queremos optimizar una funcion por lo cual primero definimos la función
#### La función seria:
   $sin(1/2*(x^2)-1/4*(y^2)+3)*cos(2*x+1-e^y)$
   
#### Podría ser cualquier funcion de dos variables.

#### La declaramos en idioma python y reemplazamos x e y por th en referencia a theta una variable muy utilizada en este tipo de cosas.

```python
# decaramos una función anonima 
func = lambda th: np.sin(1/2*(th[0]**2)-1/4*(th[1]**2)+3)*np.cos(2*th[0]+1-np.e**th[1])
#La probamos con los valores de (x = 5) e (y = 3) o mas bien: (th[0] = 5) y (th[1] = 3)
print(func([5,3]))
```

#### La funcion es un plano en tres dimensiones que ondula.
#### Para poder graficar la funcion bidimensional vamos a generar un par de vectores en x e y de al menos 100 valores y que recorra de -2 a 2
Este paso no grafica la curva, sino que genera una matriz numerica cuyos valores satisfacen la ecuacion antes dada.

```python
res = 100  # Resolucion
_X = np.linspace(-2, 2, res)
_Y = np.linspace(-2, 2, res)
# Para cada conbinacion de x e y tenemos los resultados de la funcion en Z.
_Z = np.zeros((res,res))

# Ahora recorremos la funcion con cada valore de ambos vectores 
for ix, x in enumerate(_X):
    for iy, y in enumerate(_Y):
        #el primer indice lo usas para referirte a las filas y el segundo para las columnas
        _Z[iy, ix] = func([x, y])
        
```

# Graficamos
Recordar que si no sabemos como funciona una funcion de libreria ya documentada podemos hacer por ejemplo plt.contour?

```python
plt.contour(_X,_Y,_Z, 10) # Probar variar el valor 10 por uno mas grande.
plt.show
```

```python
# Esta es una grafica bidimensional con intensidades segun la altura.
plt.contourf(_X,_Y,_Z,100) # Probar variar el valor 100 por uno mas pequeño.
plt.colorbar()
plt.show
```

### Ahora vamos a incluir un punto aleatorio (random) entre -2 y 2 que aparezca en la imagen.

```python
# Graficamos
plt.contourf(_X,_Y,_Z,100)
plt.colorbar()
#--------------------------------------------------------------------------------------------------------
# Este es el punto aleatorio Theta[0] es un valor de X y Theta[1] un valor de Y
Theta = np.random.rand(2) * 4-2 # np.random.rand(Cantidad de valores) * 4-2 (para que vaya de -2 a 2)
plt.plot(Theta[0],Theta[1], "o" , c="red") # Graficamos el punto en el plot
#--------------------------------------------------------------------------------------------------------
plt.show
# Cada vez que se ejecute aparecera en un lugar diferente.
```

# Descenso del gradiente
Repetimos el codigo con el agregado de que ahora se calculara el gradiente de la coordenada x e y para generar el gradiente el cual sera restado del valor original de la coordenada multiplicada por un factor a determinar llamado 'lr' el cual 

```python
# Graficamos
# Graficamos los 100 valores en x los 100 en y, los 100*100 resultados de la funcion dividido en 100 niveles de altura.
plt.contourf(_X,_Y,_Z,100)
plt.colorbar() # Nos deja la grafica como una imagen.
#----------------------------------------------------------------------------------------------------------

#Generamos un valor de x y y random a los que llamamos Theta[0] y Theta[1] respectivamente.
Theta = np.random.rand(2) * 4-2 # np.random.rand(Cantidad de valores) * 4-2 (para que vaya de -2 a 2)

plt.plot(Theta[0],Theta[1], "o" , c="blue") # graficamos el primer valor en blanco

#----------------------------------------------------------------------------------------------------------
# Parametros para el descenso del gradiente.
h = 0.0001 # Distancia pequeña que simula un limite tendiendo a 0 (se usa para derivar)
lr = 0.001 #learning rate o ratio de aprendizaje (el tamaño del paso en direccion a donde apunta la derivada.)
#----------------------------------------------------------------------------------------------------------
grad = np.zeros(2) # Indicara a que direccion debera dirigirse para descender
#----------------------------------------------------------------------------------------------------------
for _ in range(1000): # Iterar mil veces para que el punto de mil pasos
    
    for it, th in enumerate(Theta): #'enumerate' da los elementos de una lista mientras los enumera.it=iteracion, th = Theta
    
        _T = np.copy(Theta) # Copiamos la direccion del valor al azar
        #=================================================================================
        # Derivada de Theta por definicion
        _T[it] = _T[it] + h
        deriv = (func(_T)-func(Theta))/h #Hacemos la derivada  primero por x y despues por Y
        grad[it] = deriv # Colocamos las derivadas en las componentes del gradiente.
        #=================================================================================
    #Le restamos a la ubicacion de theta un ratio pequeño en la direccion del gradiente
    Theta = Theta - lr * grad # resta vectorial
    #-------------------------------------------------------------------------------------------------------
#    print(func(Theta)) #por si queremos ver los valores de la funcion en Theta para ver si desciende.
    #-------------------------------------------------------------------------------------------------------
    # Para poder ver el recorrido:
    if(_ % 50 == 0): # cada 100 iteraciones
        plt.plot(Theta[0],Theta[1], "." , c="red") #Para el recorrido usamos puntos y no circulos.
    #-------------------------------------------------------------------------------------------------------
plt.plot(Theta[0],Theta[1], "o" , c="green") #Coordenadas del ultimo punto en verde. 
plt.show # Graficamos
```

learning rate o ratio de aprendizaje es no solo un parametro del programa, sino que es un parametro de la red neuronal, estos son llamados mas comunmente como Hiper-Parametros.
