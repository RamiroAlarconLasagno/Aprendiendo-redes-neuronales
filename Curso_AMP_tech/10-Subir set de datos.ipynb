## Pandas es quien nos permite ingresar nuestro propio set de datos.
Para eso debemos usar el exel para ingresar los datos y luego guardarlo como <b>.csv (delimitado por comas)<b/>

```python
import pandas as pd #Pandas es quien nos permite usar nuestro propio set de datos.
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
reg = linear_model.LogisticRegression()
```

```python
# En una variable guardo la direccion y el nombre de el archivo.csv
archivo = "10-Subir set de datos propio.csv"
```

```python
df = pd.read_csv(archivo) # Cargo el contenido del archivo.
```

```python
print(df) #Mostramos los datos ingresados
df.shape #vemos las dimensiones de los datos ingresados.
```

```python
arreglox = df[df.columns[:-1]].values #Para versiones viejas .as_matrix()
arregloy = df[df.columns[-1]].values
```

```python
arreglox
```

```python
arregloy
```

### Una breve demostracion de que se pueden usar los arreglos de la misma manera que con los descargados:

```python
X_train, X_test,Y_train,Y_test = train_test_split(arreglox,arregloy)
```

```python
reg.fit(X_train,Y_train)
```

```python
reg.score(X_test,Y_test) #Los datos son muy pocos, y probablemente esta sobreentrenado, pero lo importante es que entrena.
```
